# Outliers Treatment

```python
housing_df.dtypes.value_counts()
```

```python
def get_numeric_columns(df):
    num_cols = list(df.columns[df.dtypes != 'object'])
    year_columns = ['YearBuilt_Old', 'YearRemodAdd_Old', 'GarageYrBlt_Old', 'YrSold_Old']
    for year_col in year_columns:
        if year_col in num_cols:
            num_cols.remove(year_col)
    return num_cols
```

```python
numeric_columns = get_numeric_columns(housing_df)
```

```python
numeric_columns
```

```python
for i in numeric_columns:
    data = housing_df[i]
    plt.figure(figsize=(5, 3))
    sns.boxplot(data)
    plt.title('Box Plot for ' + str(i))
    plt.show()
    print("Data differ as per the different quantile values")
    print(housing_df[i].describe([0.25,0.50,0.75,0.80,0.90,0.95,0.97,0.99]))
```

```python
outliers_percentage={}

for feature in numeric_columns:
    IQR=housing_df[feature].quantile(.75)-housing_df[feature].quantile(.25)
    outliers_count=housing_df[(housing_df[feature]>(housing_df[feature].quantile(.75)+1.5*IQR)) | (housing_df[feature]<(housing_df[feature].quantile(.25)-1.5*IQR))].shape[0]
    outliers_percentage[feature]=round(outliers_count/housing_df.shape[0]*100,2)
    
outlier_df=pd.DataFrame({'Features':list(outliers_percentage.keys()),'Percentage':list(outliers_percentage.values())})
outlier_df.sort_values(by="Percentage", ascending=False)
```

**Observations:**
- Majority of the numeric features  have outliers
- We can not drop all the outliers because it will cause loss of information.
- Hence reassigning fixed minimum and maximum values to those rows where feature value is outside the range of **[25th percentile - 1.5 * IQR, 75th percentile + 1.5 * IQR]**
- IQR or Inter Quartile Range = Difference between 75th percentile and 25th percentile values of a feature.
- Target column 'SalePrice' is excluded in this

```python
def drop_outliers(x):
    x_list = []
    for col in numeric_columns:
        if col != 'SalePrice':
            Q1 = x[col].quantile(.25)
            Q3 = x[col].quantile(.99)
            IQR = Q3 - Q1
            x = x[(x[col] >= (Q1-(1.5*IQR))) & (x[col] <= (Q3 + (1.5*IQR)))]
    return x
housing_df = drop_outliers(housing_df)
```

# Using RFE Technique for feature elemination

### Here in RFE for feature selection we are selecting multiple set of number of feature. After that we are going to choose features which we want in our final model.

```python
""" Creating a function named "top_n_features_rfe" which take "dataframe" and "no_of_features" 
which we want to select and provide top features selected by RFE """

def top_n_features_rfe(df, no_of_features):

    lm = LinearRegression()
    rfe = RFE(lm, n_features_to_select=no_of_features)
    rfe.fit(X_train, y_train)

    rfe_scores = pd.DataFrame(list(zip(X_train.columns,rfe.support_,rfe.ranking_)))
    rfe_scores.columns = ['Column_Names','Status','Rank']

    # Let's create a list of RFE selected columns.
    rfe_sele_columns = list(rfe_scores[rfe_scores.Status == True].Column_Names)
    
    return rfe_sele_columns
    
```

```python
""" Creating a function named "build_regression_model" which take "X_train", "Y_train" and "no_of_features" selected by
rfe and provide statsmodel summary."""
def build_regression_model(X_train, y_train, cols):
    X_train_sm = sm.add_constant(X_train[cols])
    lr_model = sm.OLS(y_train.values.reshape(-1, 1), X_train_sm).fit()
    
    # Let's predict y_train values using X_train_sm.
    y_train_pred = lr_model.predict(X_train_sm)
    print(lr_model.summary())
    print()
    # Let's calculate the error terms.
    #Residual Sum of Squares = Mean_Squared_Error * Total number of datapoints
    generate_model_statistics(y_train, y_train_pred)
    return lr_model
```

```python
""" Creating a function named "generate_model_statistics" which take "Y_actual" and "Y_pred" And print the 
model statistics R2 Score, RSS, MSE, RMSE value.
."""
def generate_model_statistics(y_actual, y_pred):
    # Let's print r2 score of the model.
    print("R2 Score :-", r2_score(y_actual, y_pred))

    #Residual Sum of Squares = Mean_Squared_Error * Total number of datapoints
    rss = np.sum(np.square(y_actual - y_pred))
    print("RSS : ",rss)
    mse = mean_squared_error(y_actual, y_pred)
    print("MSE : ",mse)
    rmse = mse**0.5
    print("RMSE : ",rmse)
```

We calculate the absolute value of coefficients because it allows us to treat both positive and negative coefficients with equal significance, indicating their respective relationships with the target variable, Sale price. Consider the scenario where we identify the top 10 coefficients; in this context, a coefficient of 0.00 would hold more importance than a variable with a coefficient of -0.1. However, in reality, the variable with a -0.1 coefficient is more crucial for predicting Sale price compared to the one with a coefficient of 0.00. The reason being, a coefficient of 0.00 signifies that the variable can be excluded as it is insignificant, whereas a -0.1 coefficient indicates the variable is necessary and plays a role in predicting Sale price.
